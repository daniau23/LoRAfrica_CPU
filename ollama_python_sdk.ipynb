```python
import ollama
```

For this to work, Ollama has to be running in the system background, Kindly check system tray

```python
model =  "LoRAfrica"
SYSTEM_PROMPT = "You are a helpful AI assistant specialised in African history which gives concise answers to questions asked."
question  = "Briefly detail the significance story of the Igbo god Amadioha?"
temperature  = 0.1
num_predict = 128
```

```python
messages = [
    {"role": "system", "content": SYSTEM_PROMPT},
    {"role": "user", "content": question}
]

options = {
    "temperature": temperature,
    "num_predict": num_predict
}
```

```python

response = ollama.chat(
    model=model,
    messages=messages,
    options=options
)
```

```python
response
```

```python
response.model_dump()
# response.model_dump_json()
```

```python
print(response["message"]['content'])
```

### **For Streaming**

```python
stream = ollama.chat(
    model=model,
    messages=messages,
    options=options,
    stream=True
)
```

```python
for chunk in stream:
    print(chunk["message"]['content'], end="", flush=True   )
```
